```python
features=['category','usdprice','calories (Kcal)','protein (g)','carbs (g)','fat (g)','iron (mg)']
```

```python
x=data[features].copy()
```

```python
x.sample(4)
```

```python
x['category'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore', dtype=int)

encoded = encoder.fit_transform(x[['category']])

new_cols = encoder.get_feature_names_out(['category'])

x.drop(columns=['category'], inplace=True)
x[new_cols] = encoded
```

```python
x.shape
```

```python
x.sample(2)
```

```python
print(encoder.categories_)
```

```python
onehot_cols = x.filter(like='category_').columns

x['category_cereals and tubers'] = (x[onehot_cols] == 0).all(axis=1).astype(int)
```

```python
x.sample(2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(x)
```

```python
X_scaled
```

```python
Scaled_IN_DataFrame=pd.DataFrame(X_scaled)
```

```python
Scaled_IN_DataFrame.sample(2)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
```

```python
for i in range (2,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    
```

```python
wcss
```

```python
plt.plot(range(2,10),wcss,marker="o")
plt.title("Elbow Method for Optimal K")
plt.xlabel('Number of Clusters')
plt.ylabel("WCSS")
plt.show()
# The best of num of group is 5
```
